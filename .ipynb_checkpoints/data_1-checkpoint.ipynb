```python
#create the dataset and the labels
files = os.listdir('./audio-cats-and-dogs/cats_dogs')

class_label = {'cat' : 0, 'dog' : 1}
label_class = {0 : 'cat', 1 : 'dog'}

y = np.ones(len(files))

mask = ['cat' in files[i] for i in range(len(files))]
cats = np.array(files)[mask]
dogs = np.array(files)[np.logical_not(mask)]

y[mask]=0

f1 = ['mean zero-crossing rate', 'mean spectral bandwidth','mean spectral centroid','mean root mean square energy ', 'tempo']
f2 = ['mean tempogram_' + str(i) for i in range(128)] 
f3 = ['mean mel spectrogram_'  + str(i) for i in range(128)]

f = f1 + f2 + f3

X = []
for i,file in enumerate(files):
    x,sr=librosa.load('./audio-cats-and-dogs/cats_dogs/'+file, res_type='kaiser_fast',sr=None)    
    #mel spectrum
    S=librosa.feature.melspectrogram(y=x, sr=sr, n_mels=128,fmax=8000)
    S_dB = librosa.power_to_db(S, ref=np.max)
    mean_mel_spectra = np.mean(S_dB, axis=1)

    #tempo gram
    oenv = librosa.onset.onset_strength(y=x, sr=sr, n_mels=128,fmax=8000)
    tempogram = librosa.feature.tempogram(onset_envelope=oenv, sr=sr, win_length=128)
    mean_tempogram =  np.mean(tempogram, axis=1)
    # Compute global onset autocorrelation
    ac_global = librosa.autocorrelate(oenv, max_size=tempogram.shape[0])
    ac_global = librosa.util.normalize(ac_global)
    # Estimate the global tempo for display purposes
    tempo = np.array(librosa.beat.tempo(onset_envelope=oenv, sr=sr,hop_length=128)[0]).reshape((1,))

    #mean spectral bandwith
    mean_spectral_bandwith=np.array(np.mean(librosa.feature.spectral_bandwidth(y=x,sr=sr))).reshape((1,))
    
    #mean spectral centroid 
    mean_spectral_centroid = np.array(np.mean(librosa.feature.spectral_centroid(y=x, hop_length=128))).reshape((1,))

    #mean zero-cossrate
    mean_zero_coss_rate = np.array(np.mean(librosa.feature.zero_crossing_rate(y=x))).reshape((1,))

    #mean root mean square energy 
    mean_root_mean_square_energy = np.array(np.mean(librosa.feature.rms(y=x, hop_length=128))).reshape((1,))
    
    X.append(np.concatenate((mean_zero_coss_rate,mean_spectral_bandwith,mean_spectral_centroid,mean_root_mean_square_energy,tempo, mean_tempogram, mean_mel_spectra)))

X = pd.DataFrame(columns=f, data=X)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import confusion_matrix
from mpl_toolkits.axes_grid1 import make_axes_locatable

def conf_matrix(y,y_pred, figsize=(8,8)):
    cm=confusion_matrix(y, y_pred)

    fig,ax=plt.subplots(1,1,figsize=figsize)
    mp=ax.imshow(cm.T)
    fig.colorbar(mp)
    u = np.unique(y)
    ax.set_xticks(np.arange(len(u)))
    ax.set_xticklabels(u)

    ax.set_yticks(np.arange(len(u)))
    ax.set_yticklabels(u)


    ax.set_xticks(np.arange(-.5, len(u), 1), minor=True)
    ax.set_yticks(np.arange(-.5, len(u),1), minor=True)

    ax.grid(which='minor', color='k', linestyle='-', linewidth=4)
    ax.tick_params(labelsize=15, direction='inout', size=10, width=2)

    ax.set_ylabel('Real label', fontsize=20)
    ax.set_xlabel('Predicted label', fontsize=20)

    for i in np.arange(0,len(u),1):
        for j in np.arange(0,len(u),1):
            ax.text(i,j,str(cm[i,j]),ha="center", va="center", color="k", fontsize=20)
    plt.show()
    
    
def conf_matrix2(y,y_pred,fig,ax):
    
    cm=confusion_matrix(y, y_pred)
    mp=ax.imshow(cm.T, vmin=np.min(cm), vmax=np.max([cm[i][i] for i in range(len(cm))]))
    divider = make_axes_locatable(ax)
    cax = divider.append_axes("right", size="5%", pad=0.05)
    fig.colorbar(mp, ax=ax, cax=cax)
    ax.set_xticks([0,1])
    ax.set_xticklabels(['cat', 'dog'], fontsize=15)

    ax.set_yticks([0,1])
    ax.set_yticklabels(['cat', 'dog'], fontsize=15)


    ax.set_xticks(np.arange(-.5, 2, 1), minor=True)
    ax.set_yticks(np.arange(-.5, 2,1), minor=True)

    ax.grid(which='minor', color='k', linestyle='-', linewidth=4)
    ax.tick_params(labelsize=15, direction='inout', size=10, width=2)

    ax.set_ylabel('Real label', fontsize=20)
    ax.set_xlabel('Predicted label', fontsize=20)

    for i in np.arange(0,2,1):
        for j in np.arange(0,2,1):
            ax.text(i,j,str(cm[i,j]),ha="center", va="center", color="k", fontsize=20)

    
def ACC(y_pred,y):
    return np.sum(y==np.round(y_pred)) / y.size
```

```python
#ROC curve
from sklearn.metrics import roc_curve
from sklearn.metrics import roc_auc_score

fig,ax = plt.subplots(1,1,figsize=(10,10))

y_pred=tree.predict_proba(X_test)
fpr,tpr,_ = roc_curve(y_test,y_pred[:,1])
fpr0,tpr0=fpr,tpr
ax.plot(fpr,tpr,'--',lw=3,color='tab:red', label='Tree, AUC:'+str(np.round(roc_auc_score(y_test,y_pred[:,1]),3)))

y_pred=forest.predict_proba(X_test)
fpr,tpr,_ = roc_curve(y_test,y_pred[:,1])
ax.plot(fpr,tpr,':', lw=3,color='tab:blue', label='Random forest, AUC:'+str(np.round(roc_auc_score(y_test,y_pred[:,1]),3)))

y_pred=model.predict(X_scaled_test)
fpr,tpr,_ = roc_curve(y_test,y_pred)
ax.plot(fpr,tpr,'-.', lw=3, color='tab:green', label='Neutral Network, AUC:'+str(np.round(roc_auc_score(y_test,y_pred),3)))

ax.plot([0,1],[0,1],'k', ls='-')

ax.set_xlim(-0.05, 1.05)
ax.set_ylim(-0.05, 1.05)
ax.set_xlabel('false positive rate', fontsize=20)
ax.set_ylabel('true positive rate', fontsize=20)
ax.tick_params(labelsize=15,direction='inout', size=10,width=2)
ax.legend(fontsize=20)
ax.grid()
plt.savefig("C:\\Users\\attil\\MyDocumetns\\ELTE\\Data_mining_and_machine_learning_course\\project\\latex\\fig\\ROC.pdf", bbox_inches='tight',pad_inches=0.1,dpi=500)
plt.show()
```
